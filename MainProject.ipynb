<div style="font-size:20px; margin-top:20px;">
    Step 1: Handling missing values
</div>
<div style="font-size:17px; margin-top:20px;">
    1.1 Checking missing values
</div>
<div style="font-size:17px; line-height:1.6; text-align:justify;">
Before performing any data analysis, it is important to check the dataset contains missing or null values.We can use isnull(),sum() function in Pandas to count how many missing value each column contains.

```python
df.isnull().sum()
```

```python
# Finding columns that have missing values
missing_vals = df.isnull().sum()

# Filter only columns with missing values
missing_columns=missing_vals[missing_vals > 0]

# Show results
print("Missing values:\n", missing_columns)
```

<div style="font-size:20px; margin-top:20px;">
    Step 2: Handling duplicate values
</div>
<div style="font-size:17px; line-height:1.6; text-align:justify;">
Duplicate records are identified using df.duplicated(). We can count them, display them, and remove them using df.drop_duplicates(). This helps ensure data consistency and prevents incorrect analysis results.
</div>

```python
# Remove duplicates
print("Duplicate rows before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicate rows after:", df.duplicated().sum())
print("Shape after removing duplicates:", df.shape)
```

<div style="font-size:20px; margin-top:20px;">
    Step 3: Cleaning Text Columns
</div>
<div style="font-size:17px; line-height:1.6; text-align:justify;">
    When working with text data it is important to clean and standardize the text.Cleaning these data ensure data uniformity and accurate analysis

</div>

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(r'\s+', ' ', regex=True)
df.columns = df.columns.str.replace("/", " / ")       # keep slashes
df.columns = df.columns.str.title()                   # optional
df.columns
```

```python
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].str.replace(r'\s+', ' ', regex=True)  # space cleanup only
```

```python
for col in df.columns:
    if df[col].dtype == 'object' and col != 'Target':
        continue
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
print("Final Shape:", df.shape)
```

```python
# 1. Target Variable Distribution
# Understanding the balance between Dropout/Graduate/Enrollment using Piechart

import matplotlib.pyplot as plt

# Count value for each category
target_counts=df['Student_Status'].value_counts()

# Create explode lists for correct length
explode = [0.05] * len(target_counts)

# Create Pie Chart
plt.figure(figsize=(4,4))
plt.pie(
    target_counts,
    labels=target_counts.index,
    startangle=90,
    explode= explode,
    shadow=True,
    autopct='%1.1f%%'
)

# Display 
plt.title(" Student's Status Distribution") 
plt.show()
```

```python
# 2. Academic Performance Visualisations
#  Box Plot of First Semester Grade

import seaborn as sns
import matplotlib.pyplot as plt

# Plot 
plt.figure(figsize=(7,4))
sns.boxplot(x=df['1st_Sem_Average_Grade'],color='green')

# Display
plt.title("Boxplot of 1st Semester Grades", fontsize=10)
plt.xlabel("Grade")
plt.show()
```
